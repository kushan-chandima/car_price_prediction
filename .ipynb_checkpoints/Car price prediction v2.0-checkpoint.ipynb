```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
df = pd.read_csv('car data.csv')
```

```python
df.head(5)
```

```python
df.info()
```

```python
df['Fuel_Type'].value_counts()
```

```python
df['Seller_Type'].value_counts()
```

```python
df['Transmission'].value_counts()
```

```python
df['Owner'].value_counts()
```

```python
df['Fuel_Type'].replace({'CNG':0,'Diesel':1,'Petrol':2}, inplace=True)

df['Seller_Type'].replace({'Dealer':0,'Individual':1}, inplace=True)

df['Transmission'].replace({'Manual':0,'Automatic':1}, inplace=True)
```

```python
df.head()
```

```python
df.info()
```

```python
df.corr()
```

```python
corrmat = df.corr()
top_corr_features = corrmat.index
plt.figure(figsize=(10,10))

g=sns.heatmap(df[top_corr_features].corr(),annot=True,cmap="RdYlGn")
```

```python
df.corr()['Selling_Price'].sort_values()
```

```python
x = df.drop(['Car_Name','Selling_Price'], axis=1)
y = df['Selling_Price']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1)
```

```python
print('x_train number of data: ',x_train.shape[0])
print('x_test number of data: ',x_test.shape[0])
```

```python
lre = LinearRegression()
```

```python
lre.fit(x_train[['Present_Price']],y_train)
```

```python
yhat_slr_train = lre.predict(x_train[['Present_Price']])
```

```python
lre.score(x_train[['Present_Price']],y_train)
```

```python
R2_score = r2_score(y_train,yhat_slr_train)
print(R2_score)
```

```python
ax1 = sns.distplot(y_train, hist=False, color='r', label='Actual')
sns.distplot(yhat_slr_train, hist=False,color='b', label='Predicted', ax=ax1)
plt.legend()
plt.xlabel('Selling Price')
plt.title('Actual vs Predicted')
plt.show()
```

```python
yhat_slr_test = lre.predict(x_test[['Present_Price']])
```

```python
lre.score(x_test[['Present_Price']],y_test)
```

```python
R2_score_test = r2_score(y_test,yhat_slr_test)
print(R2_score_test)
```

```python
ax1 = sns.distplot(y_test, hist=False, color='r', label='Actual')
sns.distplot(yhat_slr_test, hist=False,color='b', label='Predicted', ax=ax1)
plt.legend()
plt.xlabel('Selling Price')
plt.title('Actual vs Predicted')
plt.show()
```
